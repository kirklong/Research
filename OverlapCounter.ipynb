```julia
using Formatting
function overlapCounter(powList)
    counter1=0
    counter5=0
    errorCount=0
    iterCounter=0
    for i in 1:length(powList.ra[:])
        print(format(i/length(powList.ra[:])*100,precision=2),"% complete\r") #output % tracker
        for j in (i+1):length(powList.ra[:])
            iterCounter+=1 #for error analysis
            try
                rdist=DJM.gcdist(powList.ra[j],powList.dec[j],powList.ra[i],powList.dec[i]) #sep in deg
                if (rdist*3600)>1 && (rdist*3600)>5 #sep >5 arcsec & >1
                    counter5+=1
                elseif (rdist*3600)>1 #sep just >1 arcsec
                    counter1+=1 #sep >1 arcsec
                end
            catch
                errorCount+=1
            end
        end
    end
    println("finished with ",errorCount," errors.")
    println(format((errorCount/iterCounter)*100),"% error.")
    return counter1,counter5
end
```

```julia
overlap12_1,overlap12_5=overlapCounter(pows12)
```
